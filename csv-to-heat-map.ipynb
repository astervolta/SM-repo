### Input file name:

```python
file_in = input('CSV file name: ')
```

```python
df_ssma = pd.read_csv(file_in, header=None)
df_ssma
```

```python
month = input('input month: ')
```

```python
# HeatMap
fig, ax = plt.subplots(figsize=(14,7))         # Sample figsize in inches
heat_map = sns.heatmap(df_ssma, xticklabels=False, yticklabels=False, vmin=-1e-5, vmax=3e-5, cmap="PiYG")

# Add Title
plt.title( 'Time Averaged Map of Surface Soil Moisture Anomalies 10KM. \n NASA_USDA/HSL/SMAP10KM_soil_moisture \n over 2020-'+month+' \n')

# Display
plt.show()
```
