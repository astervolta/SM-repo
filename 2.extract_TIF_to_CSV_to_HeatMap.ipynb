### CONT. TO PLOT HEAT MAP OF INITIAL CSV

Confirm it works by plotting heat map:

```python
df_ssma = pd.read_csv('geemap_average_SMAP_SSMA_december2020.csv', header=None)
df_ssma
```

```python
# HeatMap
fig, ax = plt.subplots(figsize=(14,7))         # Sample figsize in inches
sns.heatmap(df_ssma)

# Add Title
plt.title( "Time Averaged Map of Surface Soil Moisture Anomalies SMAP DECEMBER 2020")

# Display
plt.show()
```
