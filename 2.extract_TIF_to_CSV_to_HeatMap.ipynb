```python
################################################################
# Purpose: Read TIF satellite data 
# Author: Aster Volta
# Date: 7/22/22
################################################################

# Set up the libraries and modules
#
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

import rasterio
import seaborn as sns
```

### Input file name:

```python
month = input('input month: ')
```

```python
file_in = 'geemap_average_SMAP_SSMA_'+month+'2021.tif'
```

```python
with rasterio.open(file_in) as src:
    array = src.read(1)
```

```python
array[array==0] = np.nan
array
```

## write file name below:

```python
filename_out = 'geemap_average_SMAP_SSMA_'+month+'2021.csv'
```

```python
np.savetxt(str(filename_out), array, delimiter=",")
```

### CONT. TO PLOT HEAT MAP OF INITIAL CSV

Confirm it works by plotting heat map:

```python
df_ssm = pd.read_csv(filename_out, header=None)
df_ssm
```

```python
# HeatMap
fig, ax = plt.subplots(figsize=(14,7))         # Sample figsize in inches
heat_map = sns.heatmap(df_ssm, xticklabels=False, yticklabels=False, vmin=-3, vmax=3) 
                       #cbar_kws={'label': '$mm$'}) #cmap="PiYG"

# Add Title
plt.title( 'Time Averaged Map of Surface Soil Moisture Anomalies 10KM. \n NASA_USDA/HSL/SMAP10KM_soil_moisture \n over 2021-'+month+' \n')

# Display
plt.show()
```
