##### BAND => SURFACE SOIL MOISTURE

### Input file name:

```python
month = input('input month: ')
```

```python
file_in = 'geemap_average_SMAP_SSM_'+month+'2020.csv'
```

```python
df_SSM = pd.read_csv(file_in, header=None)
df_SSM
```

```python
# HeatMap
fig, ax = plt.subplots(figsize=(14,7))         # Sample figsize in inches
heat_map = sns.heatmap(df_SSM, xticklabels=False, yticklabels=False, vmin=0, vmax=26, 
                       cbar_kws={'label': '$mm$'}) #cmap="PiYG"

# Add Title
plt.title( 'Time Averaged Map of Surface Soil Moisture 10KM. \n NASA_USDA/HSL/SMAP10KM_soil_moisture \n over 2020-'+month+' \n')

# Display
plt.show()
```
